```python
import pandas as pd
```

```python
df = pd.read_parquet('/basicsurvey.parquet')
```

```python
n_cols = []
for i in list(df.columns):
  if i[:5]!='code:' and i!='What was your biological sex assigned at birth?':
    n_cols.append(i)
```

```python
rename_cols_dict = {'In what country were you born?':'born_country',
       'Which categories describe you? Select all that apply. Note, you may select more than one group.':'race',
       'What terms best express how you describe your gender identity? (Check all that apply)':'gender',
       'Which of the following best represents how you think of yourself.':'sexual_orientation',
       'What is the highest grade or year of school you completed?':'education',
       'Have you ever served on active duty in the United States Armed forces, either in the regular military or in a National Guard or military reserve unit?':'active_duty',
       'What is your current marital status?':'marital_status',
       'Are you covered by health insurance or some other kind of health care plan?':'insurance_coverage',
       'Are you deaf, or do you have serious difficulty hearing ':'deaf',
       'Are you blind, or do you have serious difficulty seeing, even when wearing glasses':'blind',
       'Because of a physical, mental, or emotional condition, do you have serious difficulty concentrating, remembering, or making decisions':'mental_difficulty',
       'Do you have serious difficulty walking or climbing stairs':'difficulty_walking',
       'Do you have difficulty dressing or bathing':'difficulty_dressing',
       'Because of a physical, mental, or emotional condition, do you have difficulty doing errands alone such as visiting a physician\'s office or shopping':'difficulty_errands',
       'What is your current employment status? Please select 1 or more of these categories.':'employment',
       'What is your annual household income from all sources?':'income',
       'Do you own or rent the place where you live?':'own_or_rent',
       'How many years have you lived at your current address?':'years_in_current_address',
       'In the past 6 months, have you been worried or concerned about NOT having a place to live?':'stable_house_concern'}
```

```python
df_new = df[n_cols]
```

```python
df_new.rename(columns=rename_cols_dict, inplace=True)
```

```python
df_new.to_parquet('/basicsurveycleaned.parquet')
```
